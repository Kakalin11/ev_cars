### Most of the data has units that is unnecessary to have, let's get rid of them

```python
df['Accel']=df['Accel'].str.split('sec',expand=True)[0]
df['Accel_sec']=pd.to_numeric(df['Accel'])
df = df.drop(['Accel'], axis=1)
```

```python
df['TopSpeed']=df['TopSpeed'].str.split('km/h',expand=True)[0]
df['TopSpeed_km/h']=pd.to_numeric(df['TopSpeed'])
df = df.drop(['TopSpeed'], axis=1)
```

```python
df['Range']=df['Range'].str.split('km',expand=True)[0]
df['Range_km']=pd.to_numeric(df['Range'])
df = df.drop(['Range'], axis=1)
```

```python
df['Efficiency']=df['Efficiency'].str.split('Wh/km',expand=True)[0]
df['Efficiency_Wh']=pd.to_numeric(df['Efficiency'])
df = df.drop(['Efficiency'], axis=1)
```

```python
df['FastCharge']=df['FastCharge'].replace('-','0 km/h') #There is a non-integer value in the data that needs to be converted
```

```python
df['FastCharge']=df['FastCharge'].str.split('km/h',expand=True)[0]
df['FastCharge_km/h']=pd.to_numeric(df['FastCharge'])
df = df.drop(['FastCharge'], axis=1)
```

```python
df.info()
```

```python
df.head(10)
```

### Continue to clean and organize data 

```python
df['PowerTrain'].unique() #Let's replace the values with their commanly used abbrevations
```

```python
df['PowerTrain']=df['PowerTrain'].replace('All Wheel Drive','AWD')
df['PowerTrain']=df['PowerTrain'].replace('Rear Wheel Drive','RWD')
df['PowerTrain']=df['PowerTrain'].replace('Front Wheel Drive','FWD')
```

```python
df['RapidCharge'].unique() # Since it has two values, let's call them yes and no
```

```python
df['RapidCharge'] = df['RapidCharge'].replace('Rapid charging possible','Yes')
df['RapidCharge'] = df['RapidCharge'].replace('Rapid charging not possible','No')
```
